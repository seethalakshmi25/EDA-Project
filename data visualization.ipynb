```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Results.csv')
df.head()
```

```python
column_names = ['company', 'location', 'industry', 'total_laid_off', 'percentage_laid_off', 'stage', 'country', 'funds_raised_millions', 'date']

df = pd.read_csv('Results.csv', header=None, names=column_names)
df.head()
```

```python
industry_totals = df.groupby('industry')['total_laid_off'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=industry_totals.values, y=industry_totals.index, palette='viridis')
plt.xlabel('Total Laid Off')
plt.ylabel('Industry')
plt.title('Top 10 Industries by Total Layoffs')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.to_period('M')

monthly_totals = df.groupby('month')['total_laid_off'].sum()

plt.figure(figsize=(14,6))
monthly_totals.plot(kind='line', marker='o')
plt.xlabel('Month')
plt.ylabel('Total Laid Off')
plt.title('Monthly Layoffs Trend (2020-2023)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
country_totals = df.groupby('country')['total_laid_off'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=country_totals.values, y=country_totals.index, palette='mako')
plt.xlabel('Total Laid Off')
plt.ylabel('Country')
plt.title('Top 10 Countries by Total Layoffs')
plt.show()
```

```python
stage_totals = df.groupby('stage')['total_laid_off'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=stage_totals.values, y=stage_totals.index, palette='rocket')
plt.xlabel('Total Laid Off')
plt.ylabel('Stage')
plt.title('Total Layoffs by Company Stage')
plt.show()
```
